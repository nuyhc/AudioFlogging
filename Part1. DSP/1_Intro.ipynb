## 1-1. 신호의 수학적 표현
> 신호들은 변화의 패턴으로, 정보를 표현하거나 부호화 함

많은 신호들이 시간축에서 변환하고 우리가 흔히 아는, 시간 파현(Time waveform)을 만든다.

위 파형처럼, 연속시간 신호(continuous-time signal)은, 보통 $t$로 표현되는 독립변수의 함수로 수학적 표현이 가능하다.  
그래프는 단순한 식으로 쓸 수 없음에도 불구하고 함수 $s(t)$를 사용하는데, 실제로 그래프 그 자체는 각 순간 시각 $t$에 해당하는 $s(t)$값으로 정의 될 수 있기 때문이다.  

대다수의 많은 신호들이 연속시간 신호들로 발생되지만, 시간축상 일정한 간격으로 연속시간 신호를 **샘플링(sampling)** 함으로써 이산시간 표현을 구할 수 있다.  
이산값만 갖는 인덱스 변수의 함수로 표현되는 수열이 생성되는데,  
$$s[n] = s(nT_s), \ n=\{..., -1, 0, 1, ...\}$$  
와 같이 표현할 수 있고, $T_s$는 **샘플링 주기(sampling period)** 이다.
