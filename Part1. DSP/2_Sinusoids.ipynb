# Part2. 정현파 (Sinusoid)
> 사인(sine) 또는 코사인(cosine) 신호: 정현파 신호(sinusoidal signal)  

수학적으로 간단하게 표현되지만, 신호와 시스템에서 가장 근본적인 신호이므로, 반드시 숙지해야하는 신호의 형태이다.  
코사인 신호의 일반적인 형태는 다음과 같다.  
$$x(t) = Acos(\omega_0t+\phi)$$  
연속시간 신호를 정의할 때, 우리는 독립 변수가 $t$인 함수를 주로 사용하는데, 이는 시간을 표현하는 연속변수이다.  
  
$t$와 $cos$을 제외한, $A, \omega_0, \phi$는 각각  
진폭(Amplitude), 라디안 주파수(radian freq.), 위상 변이(phase shift)를 의미한다.

해당 신호의 표현을 해석하면,  
$x(t)$는 -A ~ A 사이에서 진동하고, $1 \over \omega_0$의 주기(period)를 갖는다고 해석할 수 있다.

### 사인 코사인 함수 (삼각함수)

```python
x = np.linspace(-3*np.pi, 3*np.pi, 100)
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)
ax[0].plot(x, np.sin(x))
ax[0].set_title("Sin")
ax[1].plot(x, np.cos(x))
ax[1].set_title("Cos")
for idx in [0, 1]: ax[idx].axhline(y=0, color="r", linewidth=1, linestyle=":")
plt.show()
```

정현파 신호는 우리가 익히 알고 있는 삼각함수의 사인과 코사인 함수로 정의되는데, 삼각함수는 매개변수가 각도이다.  
각도는 도(degree)로 표현되지만, 사인과 코사인의 경우 각도의 차원이 존재하지 않으므로 라디안(radian)으로 표현한다.  

두 함수 모두 -1 ~ +1 사이에서 진동하고 $2\pi$ 주기로 같은 형태가 반복된다.  
사인 함수의 경우, 좌우가 대칭이 아닌 기함수(odd)이고 코사인 함수의 경우, 좌우가 대칭인 우함수(even)이다.  

### 특성

| 특성 | 식 |
| :---: | :---: |
| 동질성 | $sin\theta = cos(\theta-\pi/2)$ <br> $cos\theta = sin(\theta+\pi/2)$ |
| 주기성 | $cos(\theta+2\pi k) = cos\theta$ |
| 우함수 | $cos(-\theta)=cos\theta$ |
| 기함수 | $sin(-\theta)=-sin\theta$ |

### 삼각함수 공식
- $sin^2\theta + cos^2\theta = 1$
- $cos2\theta = cos^2\theta - sin^2\theta$
- $sin2\theta = 2sin\theta cos\theta$
- $sin(\alpha \pm \beta) = sin\alpha cos\beta \pm cos\alpha sin\beta$
- $cos(\alpha \pm \beta) = cos\alpha cos\beta \mp sin\alpha sin\beta$

## 2-1.정현파 신호
> 코사인의 매개변수를 시간의 함수로 정의하면, 가장 일반적인 정현파 신호가 된다.  

$$x(t)=Acos(\omega_0t+\phi)=Acos(2\pi f_0t+\phi)$$
1. $A$는 진폭(amplitude)이라고 한다. 진폭은 코사인 함수의 진동 범위(크기)를 정하는 인자.
2. $\phi$는 위상 변이(phase shift)라고 한다. 위상 변이의 단위는 라디안으로, 일반적으로 위상 변이는 코사인으로 정의하는 것을 선호함.  
   $x(t)=Asin(\omega_0t+\phi')=Acos(\omega_0t+\phi'-\pi/2)$
3. $\omega_0$는 라디안 주파수(radian frequency)이다. 

```python
def x(t):
    return 20*np.cos(2*np.pi*40*t-0.4*np.pi)

t = np.linspace(-0.05, 0.05, 100)
plt.plot(t, x(t))
plt.axhline(y=0, color="r", linestyle=":")
plt.show()
```

$x(t)=20cos(2\pi(40)t-0.4\pi)$를 시각화하면 위와 같다.  
앞서 정의한 정의를 적용하면, $A=20$, $\omega_0=2\pi(40)$, $f_0=40$, $\phi=-0.4\pi$이다.

### 주파수와 주기의 관계
주기(period)는 $T_0$로 표시하며, 한 순환의 길이를 의미한다.  
일반적으로 정현파의 주파수로 그 주기를 결정할 수 있다.  
- $x(t+T_0) = x(t)$
- $Acos(\omega_0(t+T_0)+\phi) = Acos(\omega_0t+\phi)$

코사인 함수의 주기는 $2\pi$이므로, 등식은 모든 $t$에 대하여 성립한다.
- $\omega_0T_0=2\pi \rightarrow T_0={2\pi \over \omega_0}$
- $(2\pi f_0)T_0 = 2\pi \rightarrow T_0 = {1 \over f_0}$

```python
def cos_per_freq(t, f):
    return 5*np.cos(2*np.pi*f*t)

t = np.linspace(-10, 10, 1000)
fig, ax = plt.subplots(nrows=3, ncols=1, sharex=True, sharey=True)
ax[0].plot(t, cos_per_freq(t, 0))
ax[0].axhline(y=0, color="r", linestyle=":")
ax[0].set_title("f=0Hz")
ax[1].plot(t, cos_per_freq(t, 100))
ax[1].axhline(y=0, color="r", linestyle=":")
ax[1].set_title("f=100Hz")
ax[2].plot(t, cos_per_freq(t, 200))
ax[2].axhline(y=0, color="r", linestyle=":")
ax[2].set_title("f=200Hz")
plt.show()
```

주파수를 이해하는 것은 매우 중요한데, 위 그림을 통해 주파수에 대한 효과를 살펴 볼 수 있다.  
$x(t)=5cos(2\pi f_0 t)$의 함수를 주파수를 다르게해 시각화한것은 위와 같다.  

파형은 비슷하나, 고주파일수록(f=200) 신호가 시간에 따라 더 빨리 변화하는 것을 알 수 있다.  
시간에 따라 더 빨리 변화한다는 것은 주기가 더 짧다는 의미인데, 이를 통해 주파수와 주기는 반비례 관계임을 알 수 있다.  

주파수가 0인 경우에는, $5cos(2\pi \cdot 0 \cdot t)=5$이고 변화가 없는 DC라 부른다.  

### 위상 변이와 시간 변이의 관계
위상 변이 변수인 $\phi$는 주파수와 함께 코사인 함수에서 최대치와 최소치의 위치를 결정한다.  

예를 들어, $x(t)=Acos(\omega_0t+\phi)=Acos(2\pi f_0t+\phi)$에서  
$\phi=0$이면 $t=0$에서 피크가 발생한다.  
반대로, $\phi \neq 0$이면,  
피크 발생 위치가 $t=0$에서 시간적으로 얼마나 변이되는지를 결정한다.

```python
t = list(range(-3, 3))
raw = [0 for _ in t]
raw[3] = 1
delay = [0 for _ in t]
delay[4] = 1
adv = [0 for _ in t]
adv[2] = 1

fig, ax = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize=(12, 3))
ax[0].plot(t, adv)
ax[0].set_title("선행(advanced)")
ax[1].plot(t, raw)
ax[1].set_title("Original")
ax[2].plot(t, delay)
ax[2].set_title("지연(delay)")
plt.show()
```

위 이미지는, 위상변이의 이해를 돕기 위한 예시 이미지이다.  

중앙의 Original을 원본 신호 $x(t)=s(t)$라고하면,  
좌측은 $x(t)=s(t+1)$, 우측은 $x(t)=s(t-1)$이 성립하게 된다.  

시간적으로 변이가 양수로 발생하면, 오른쪽으로 변이된 것이고 이를 지연(delay)라하며,  
변이가 음수로 발생하면, 왼쪽으로 변이된 것이고 이는 선행(advanced)이라고 한다.  
즉, 변이를 통해 신호의 원점을 새롭게 재정의하는 셈이다.  

위와 유사한 방법으로, 코사인 함수에서 시간 변이를 알아내는 한가지 방법은 $t=0$에서 가장 가까운 양의 피크(최댓값)를 찾는 것이다.  

$x_0(t)=Acos(\omega_0t)$가 위상이 0인 함수라면,  
$x_0(t-t_1) = Acos(\omega_0(t-t_1))=Acos(\omega_0 t+\phi)$,  
$cos(\omega_0t-\omega_0t_1)=cos(\omega_0t+\phi)$가 성립하므로 아래 식이 성립한다.  
$$t_1 = - {\phi \over \omega_0} = - {\phi \over 2\pi f_0}$$  

지연(시간 변이가 양수)가 발생하는 경우, 위상 변이는 음수이고 주기에 대해 동일하게 적용하면 다음과 같은 식을 얻을 수 있다.  
$$\phi = -2\pi f_0 t_1 = -2\pi( {t_1 \over T_0} )$$  

원점 $t=0$에 가장 가까운 양의 피크는 언제나 구간 $|t_1| \leq T_0/2 $내에 존재하므로, 위상 변이도 항상 $-\pi < \phi < \pi$를 만족한다.  
다만, 코사인과 사인 함수는 기본적으로 $2\pi$의 주기를 갖는 주기 함수이므로, 구간에 따라 위상 변이값이 변하는 모호함이 존재한다.  
이와 같은 주기성을 이용해, $2\pi$마다 서로 다른 피크를 찾을 수 있고, 이는 정현파의 양의 피크를 하나 찾아 이에 대응하는 시간 변이값을 정하는 또 다른 방법이 될 수 있다.  

$-\pi$와 $\pi$ 사이에 위치하는 위상 변이값을 위상 변이의 **기본값(principal value)** 이라고 한다.

## 2-2. 정현파의 샘플링과 시각화
> 컴퓨터는 신호를 이산신호로 다룬다.
$$x(t)=20cos(2\pi(40)t-0.4\pi)$$
$$x(nT_s)=20cos(80\pi n T_s-0.4\pi)$$

$T_s$는 샘플링 간격(sampling spacing) 또는 샘플링 주기(sampling period)라고 한다.  

```python
def xnTs(Tn):
    return 20*np.cos(80*np.pi*Tn-0.4*np.pi)

x_ts5 = np.linspace(-0.03, 0.03, 5)
x_ts10 = np.linspace(-0.03, 0.03, 10)
x_ts20 = np.linspace(-0.03, 0.03, 20)
x_ts30 = np.linspace(-0.03, 0.03, 30)
x_ts40 = np.linspace(-0.03, 0.03, 40)
x_ts50 = np.linspace(-0.03, 0.03, 50)

fig, ax = plt.subplots(6, 1)
for idx, rng in enumerate([x_ts5, x_ts10, x_ts20, x_ts30, x_ts40, x_ts50]):
    ax[idx].plot(rng, xnTs(rng))
```

위 그래프는 샘플링의 차이에 따른 신호 형상을 그린것과 비슷하다.  
샘플링 주기가 짧아질수록, 한 주기에서 얻는 샘플의 개수가 늘어나 특정 샘플링 주기 이상이 되어야  
우리가 흔히아는 형태와 유사해진다.  

샘플링 주기($Ts$)는 주파수와 관계가 있는데, 샘플로부터 원래의 신호를 재구성하는데 샘플 간격이 주기의 반 이하가 되면 충분하다.  
해당 원리는 이후에 알아보겠지만, 나이퀴스트 이론과 관련된 내용이다.

## 2-3. 복소 지수 함수와 페이저
여러 신호를 코사인 함수 형태로 표현하는 것이 효과적이면서 해석에 용이하다는 이야기를 앞서했다.  
하지만, 정현파 신호를 분석하고 취급하는 데 복소 지수 신호(complex exponential signals)를 이용하면 더욱 간단해진다.  

### 복소수
하나의 복소수 $z$는 실수의 순서 쌍(ordered pair)으로 나타낸다.  
$z=(x,y)$로 복소수를 나타낼 수 있고, $x=\Re e\{z\}$는 실수부(real part), $y=\Im m\{z\}$는 허수부(imaginary part)이다.  

직교좌표형(Cartesian form)으로 나타내는 경우 $z=x+jy$ 형태로 표시할 수 있다.  

벡터는 길이와 방향으로 표현되므로,  
복소수를 극좌표형(polar form)으로 표현하면 벡터의 길이 $r$과 실수축과 이루는 각 $\theta$로 표현할 수 있다.  
$z=r\angle \theta$, $x=rcos\theta$, $y=rsin\theta$  
극좌표계를 표현하는데, 오일러 공식을 사용하는 것도 가능하다.  
$e^{j\theta}=cos\theta + jsin\theta$ -> $z=re^{j\theta}=rcos\theta + jrsin\theta$

만약, 극좌표계에서 직각좌표계로 바꾸러면 아래 공식을 사용하면 된다.  
$r=\sqrt{x^2+y^2}$, $\theta=arctan( {y \over x} )$

### 복소 지수 신호
$z(t)=Ae^{ j(\omega_0t+\phi) }$로 나타내는 복소 지수 신호는 다음과 같이 직각좌표계로 표현하면,  
$z(t)=Ae^{ j(\omega_0t+\phi) }=Acos(\omega_0t+\phi) + jAsin(\omega_0t+\phi)$로 나타 낼 수 있다.

### 역 오일러 공식
역 오일러 공식을 이용해 코사인 함수를 복소 지수로 나타낼 수 있다.  
$cos\theta = { e^{j\theta} + e^{-j\theta} \over 2 }$, $sin\theta = { e^{j\theta} - e^{-j\theta} \over 2j }$
