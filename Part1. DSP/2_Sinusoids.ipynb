# Part2. 정현파 (Sinusoid)
> 사인(sine) 또는 코사인(cosine) 신호: 정현파 신호(sinusoidal signal)  

수학적으로 간단하게 표현되지만, 신호와 시스템에서 가장 근본적인 신호이므로, 반드시 숙지해야하는 신호의 형태이다.  
코사인 신호의 일반적인 형태는 다음과 같다.  
$$x(t) = Acos(\omega_0t+\phi)$$  
연속시간 신호를 정의할 때, 우리는 독립 변수가 $t$인 함수를 주로 사용하는데, 이는 시간을 표현하는 연속변수이다.  
  
$t$와 $cos$을 제외한, $A, \omega_0, \phi$는 각각  
진폭(Amplitude), 라디안 주파수(radian freq.), 위상 변이(phase shift)를 의미한다.

해당 신호의 표현을 해석하면,  
$x(t)$는 -A ~ A 사이에서 진동하고, $1 \over \omega_0$의 주기(period)를 갖는다고 해석할 수 있다.

### 사인 코사인 함수 (삼각함수)

```python
x = np.linspace(-3*np.pi, 3*np.pi, 100)
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)
ax[0].plot(x, np.sin(x))
ax[0].set_title("Sin")
ax[1].plot(x, np.cos(x))
ax[1].set_title("Cos")
for idx in [0, 1]: ax[idx].axhline(y=0, color="r", linewidth=1, linestyle=":")
plt.show()
```

정현파 신호는 우리가 익히 알고 있는 삼각함수의 사인과 코사인 함수로 정의되는데, 삼각함수는 매개변수가 각도이다.  
각도는 도(degree)로 표현되지만, 사인과 코사인의 경우 각도의 차원이 존재하지 않으므로 라디안(radian)으로 표현한다.  

두 함수 모두 -1 ~ +1 사이에서 진동하고 $2\pi$ 주기로 같은 형태가 반복된다.  
사인 함수의 경우, 좌우가 대칭이 아닌 기함수(odd)이고 코사인 함수의 경우, 좌우가 대칭인 우함수(even)이다.  

### 특성

| 특성 | 식 |
| :---: | :---: |
| 동질성 | $sin\theta = cos(\theta-\pi/2)$ <br> $cos\theta = sin(\theta+\pi/2)$ |
| 주기성 | $cos(\theta+2\pi k) = cos\theta$ |
| 우함수 | $cos(-\theta)=cos\theta$ |
| 기함수 | $sin(-\theta)=-sin\theta$ |

### 삼각함수 공식
- $sin^2\theta + cos^2\theta = 1$
- $cos2\theta = cos^2\theta - sin^2\theta$
- $sin2\theta = 2sin\theta cos\theta$
- $sin(\alpha \pm \beta) = sin\alpha cos\beta \pm cos\alpha sin\beta$
- $cos(\alpha \pm \beta) = cos\alpha cos\beta \mp sin\alpha sin\beta$
